```python
#Body Mass Index
bmi=float(input("Enter the BMI Index:" ))

if(bmi<18.5):
    print("Underweight")
elif( 18.5 <=bmi<= 24.9):
    print("Normal Weight")
elif( 25.0 <=bmi<= 29.9):
    print("Overweight")
elif( 30.0 <=bmi<= 34.9):
    print("Obese (Class I)")
elif(35.0 <=bmi<= 39.9):
    print("Obese (Class II)")
else:
    print("Obese (Class III) (Morbid Obesity)")            
                


                
```
